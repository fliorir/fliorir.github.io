```python
# -*- coding: utf-8 -*-
"""
Created on Mon Jul 25 08:00:37 2016

@author: krasin
"""

import pandas as pd
reportFile = "GoogleBooksTrafficReport_2017_09.csv"
metaFile = "id_table.csv"

book_report = pd.read_csv(reportFile, sep = '\t', encoding='utf-16')
book_id = pd.read_csv(metaFile, sep = '\t', header=None, encoding='utf-8')

book_report.describe()

book_report.columns

def seriesToInt(num):
    num = num.str.replace(u'\xa0', '').astype(int)
    return num
    
 
book_report['Book Visits (BV)'] = seriesToInt(book_report['Book Visits (BV)'])
book_report['BV with Pages Viewed'] = seriesToInt(book_report['BV with Pages Viewed'])
book_report['Pages Viewed'] = seriesToInt(book_report['Pages Viewed'])
```

```python
book_report.replace(u'Пiсня про людину / Песен за човека', 
                    u'Песен за човека / Пiсня про людину',
                   inplace=True)
book_report.replace(u'В родината на Орфей', u'В родината на Орфей / На родине Орфея', inplace=True)
book_report.head(8)
```

```python
#initial_columns = book_report.columns
book_report.columns = ['isbn', 'title', 'BV', 'bookVisits', 'clicks', 'clicks2', 'ctr', 'pagesViewed']
book_report.head(6)
```

```python
book_id.columns = ['seqN','id', 'title', 'authors', 'date_print', 'date_ebook']
book_id.tail()
```

```python
b = pd.merge(book_id, book_report, how='inner', on=None,
         left_on='title', right_on='title',
      left_index=False, right_index=False, sort=False,
      copy=True, indicator=False)
b.head()
```

```python
def generateDiv(s):
    title = u'<div> <a href="http://books.google.com/books?{}&printsec=frontcover"><b>"{}"</b></a>'.format(s['id'], s['title'])
    auth = u' {} '.format(s.authors)
    inPrint = u'(книжно издание {}; електронно издание {})'.format(s.date_print, s.date_ebook)
    stats = u' - {} читатели, прочели {} страници.'.format(s['bookVisits'].astype(int), s['pagesViewed'].astype(int))
    div2 = u'</div>'
    #(книжно издание 20.10.2008; електронно издание 29.09.2008)
    return title + auth + inPrint + stats + div2
    
def getAuthor(s):
    index = s.find('[') - 1
    return s[:index]
```

```python
import datetime
now = datetime.datetime.now()
intro1 = u'<b>Общо <i>(до {day}.{month}.{year} г.)</i> '.format(day = now.day,
                                                                month = now.month, 
                                                                year = now.year)
intro2 = u'<a href="http://books.google.com/books?q=inpublisher:fliorir">електронните книги на издателство "Фльорир"</a> '
intro3 = u'са посетени от {:d} читатели, прочели {} страници.</b>\n'.format(b['bookVisits'].sum(), b['pagesViewed'].sum())
intro4 = u'<div><span style="color: red;">' + \
         u'<b>Посещаемост на електронните издания на книги на издателство "Фльорир":</b></span></div>\n\n'
intro = intro1 + intro2 + intro3 + intro4
ending = u'<span style="color: #33cc00;"><b><i>Частично позициониране на страницата - на всеки 30 дни.</i></b></span>'

with open("Output.txt", "a") as f:
    f.write((intro).encode('utf8'))
    for i in range(len(b)):
        f.write(generateDiv(b.iloc[i]).encode('utf8'))
        f.write(u'\n\n')
    f.write(ending.encode('utf8'))
```
